# Broadcasting 소개

```python
import numpy as np
```

```python
a = np.array([0, 1, 2])
b = np.array([5, 5, 5])
a + b
```

```python
M = np.ones((3, 3))
M + a # a의 차원 변형 (Broadcasting)
```

```python
a = np.arange(3)
b = np.arange(3)[:, np.newaxis]

print(a)
print(b)
```

```python
# a = (3, ), b = (3, 1)
# 1. a = (1, 3)으로 차원 변환
# 2. 크기를 맞추기 위해 늘림 -> (3, 3)
# 3. 합연산
a + b
```

### Broadcasting Visualization

```python
# Example 1

M = np.ones((2, 3))
a = np.arange(3)
```

```python
print(f"M : {M}\n")
print(f"a : {a}")
```

```python
M + a
```

```python
# Example 2
a = np.arange(3).reshape(3, 1)
b = np.arange(3)
```

```python
print(f"a = {a}")
print(f"b = {b}")
```

```python
a + b
```

```python
# Exmaple 3
M = np.ones((3, 2))
a = np.arange(3)
```

```python
print(f"M = {M}")
print(f"a = {a}")
```

```python
# 3x2와 3,
M + a
```

```python
# 3, -> 1x3으로 차원 변환
a[:, np.newaxis]
```

```python
M + a[:, np.newaxis]
```

## Centering in Array

```python
rng = np.random.default_rng(seed = 1701)
X = rng.random((10, 3)) * 100
print(X)
```

```python
Xmean = X.mean(0)
Xmean
```

```python
X_centered = X - Xmean
X_centered
```

```python
X_centered.mean(0)
```
